# ml07. K-Means

### Clustering(군집)분석::  데이터셋의 관측값이 가지고 있는 여러 속성을 분석해서 
### 서로 비슷한 특징을 갖는 관측값끼리 같은 클러스터(집단)으로 묶는 알고리즘
### 다른 클러스터간에는 서로 완전하게 구분되는 특징이 있기 때문에 어느 클러스터도
###  속하지 못하는 특징 관측값존재한다. => 데이터의 이상값, 중복값을 찾는데 활용하기도 한다.

```python
import pandas as pd
import matplotlib as plt
```

### Step1: 데이터 준비

```python
#도매업(wholesale customers) 고객 데이터 셋
#: 고객의 연간 구매금액을 상품 카테고리 별로 구분하여 정리한 데이터
uci_path='https://archive.ics.uci.edu/ml/machine-learning-databases/00292/Wholesale%20customers%20data.csv'
df=pd.read_csv(uci_path, header=0)
```

### Step2:: 데이터 탐색

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

### Step3: 데이터 전처리(Preprocessing)

```python
# 분석에 사용할 속성을 선택
X=df.iloc[:, :] #모든행 모든열
X[:5]
```

```python
# 설명변수 데이터를 정규화
from sklearn import preprocessing
```

```python
X=preprocessing.StandardScaler().fit(X).transform(X)
X[:5]
```

### Step4 : k-means 군집 모델-sklearn 사용

```python
#  sklearn 라이브러리에서 cluster 군집 모델 가져오기
from sklearn import cluster
```

```python
# 모델 객체 생성
#n_clusters=5: 클러스터의 갯수
kmeans=cluster.KMeans(init='k-means++',n_clusters=5, n_init=10)
```

```python
#model fitting(모델 학습)
kmeans.fit(X)
```

```python
# 예측(군집) : 0~4군집
cluster_label=kmeans.labels_
cluster_label
```

```python
# 예측결과를 DataFrame에 추가
df['Cluster']=cluster_label
df.head()
```

```python
# 그래프로 시각화
df.plot(kind='scatter', x='Grocery', y='Frozen', c='Cluster', 
          cmap='Set1', colorbar=False, figsize=(10,10))
df.plot(kind='scatter', x='Milk', y='Delicassen', c='Cluster', 
          cmap='Set1', colorbar=True, figsize=(10,10))
plt.show()
plt.close()
```

```python
# 클러스터 라벨이 0,4 제외 : 이상치에 가깝기 때문
# 데이터가 몰려 있는 구간을 확대: 1,2,3 클러스터
mask=(df['Cluster']==0) | (df['Cluster']==4)
ndf=df[~mask]  #cluster : 1,2,3
```

```python
# 그래프로 시각화
ndf.plot(kind='scatter', x='Grocery', y='Frozen', c='Cluster', 
          cmap='Set1', colorbar=False, figsize=(10,10))
ndf.plot(kind='scatter', x='Milk', y='Delicassen', c='Cluster', 
          cmap='Set1', colorbar=True, figsize=(10,10))
plt.show()
plt.close()
```
