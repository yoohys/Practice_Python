# Ch03-3_Linear_Regression

### y=XW+b (Y: output, X: input, b: Bias(편량))
### Y= sigmoid(XW+b) : 활성화 함수

```python
import tensorflow as tf
```

```python
x_data=[1,2,3]
y_data=[1,2,3]
```

```python
W=tf.Variable(tf.random_uniform([1],-1.0,1.0))
b=tf.Variable(tf.random_uniform([1],-1.0,1.0))
```

```python
X=tf.placeholder(tf.float32,name="X")
Y=tf.placeholder(tf.float32,name="Y")
print(X)
print(Y)
```

```python
hyypthesis=W+X+b
```

```python
# loss function (손실 함수) : 한상(x, y) 데이터에 대한 손실값 게산하는 삼수
# : 예측값에서 실제값을 뺸 뒤 제곱하는 차이값
cost=tf.reduce_mean(tf.square(hyypthesis-Y)) # 최소 제곱법
                    
# Gradiant Descent(경사하강법) : 최적화 함수, 함수의 기울기를 구하고 기울기가 낮은 쪽으로 계쏙 이동시키면서
#: 최적의 값을 찾아내는 방법

optimizer=tf.train.GradientDescentOptimizer(learning_rate=0.1)

train_op=optimizer.minimize(cost)
```

```python
# with ~as : 인터프리터가 자동으로 구문이 끝나면 닫기

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    #최적화 100번 수행 
    for step in range(100):
        # _ : 수행된 최종결과를 저장하겠따
        _, cost_val= sess.run([train_op,cost],feed_dict={X:x_data,Y:y_data})
        print(step, cost_val, sess.run(W), sess.run(b))
    
    #결과 확인
    print("\n======== Test ==========")
    print('X:5, Y: ',sess.run(hyypthesis,feed_dict={X:5}))
    print('x:2.5 Y: ',sess.run(hyypthesis,feed_dict={X:2.5}))
```
