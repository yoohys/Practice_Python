## 파이썬에서 가장많이 사용, matplotlib(기본형)
## 데이터상관관계에 특화

```python
import seaborn as sns

from matplotlib import pyplot as plt
```

## (1) iris 데이터 불러오기 (seaborn 내장이 되어있음)

```python
df=sns.load_dataset('iris')
df.head()
```

## (2) scatter plot

```python
sns.regplot(x=df["sepal_length"],y=df["petal_length"],line_kws={'color':'red'})
```

```python
sns.regplot(x=df["sepal_length"],y=df["petal_length"],line_kws={'color':'red'},scatter_kws={'color':"green"})
```

## (3) 선 없애기

```python
ax = sns.regplot(x=df["sepal_length"],y=df["petal_length"],scatter_kws={'color':'green'},fit_reg=False)
ax.set_ylim([0,8])
ax.set_xlabel("Sepallength")
```

## (4) 그림 속성 바꾸기

```python
ax = sns.regplot(x=df["sepal_length"],y=df["petal_length"])
ax.figure.set_size_inches
```

## (5) 저장

```python
sns.set_style("darkgrid")
ax=sns.regplot(x=df["sepal_length"],y=df["petal_length"])
ax.figure.savefig("iris.png")
```

## (6) Histogram & 밀도 그래프

```python
sns.distplot(df['sepal_width'],bins=4) #bins: 막대의 갯수
```

```python
sns.distplot(df['sepal_width'],bins=[2,2.5,3,3.5,4,4.5,5]) #bins = 막대의 갯수
```
